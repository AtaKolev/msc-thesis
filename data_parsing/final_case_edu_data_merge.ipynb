```python
def rename_columns(df):

    cols_to_rename = [x for x in df.columns if x[0] == 'X']
    dicti = {}
    for col in cols_to_rename:
        dicti.update({col : col.split('_')[1]})
    
    return df.rename(columns = dicti)
```

```python
final_dfs = {'1730' : pd.DataFrame(),
             '1750' : pd.DataFrame(),
             '1772' : pd.DataFrame(),
             '1797' : pd.DataFrame()}
for normal_file in glob.glob(os.path.join(normal_data_folder, '*.csv')):
    rpm_normal = normal_file.split('_')[7]
    final_df = pd.DataFrame()
    df_normal = pd.read_csv(normal_file)
    df_normal = rename_columns(df_normal)
    for de_fault_file in glob.glob(os.path.join(de_fault_12_folder, '*.csv')):
        de_fault_element = de_fault_file.split('_')[-1].split('.')[0]
        if rpm_normal in de_fault_file:
            de_fault = pd.read_csv(de_fault_file)
            de_fault = rename_columns(de_fault)
            de_fault['fault_element'] = de_fault_element
            if final_df.empty:
                final_df = pd.concat((df_normal, de_fault), axis = 0)
            else:
                final_df = pd.concat((final_df, de_fault), axis = 0)
    for fe_fault_file in glob.glob(os.path.join(fe_fault_12_folder, '*.csv')):
        fe_fault_element = fe_fault_file.split('_')[-1].split('.')[0]
        if rpm_normal in fe_fault_file:
            fe_fault = pd.read_csv(fe_fault_file)
            fe_fault = rename_columns(fe_fault)
            fe_fault['fault_element'] = fe_fault_element
            if final_df.empty:
                final_df = pd.concat((df_normal, fe_fault), axis = 0)
            else:
                final_df = pd.concat((final_df, fe_fault), axis = 0)
    final_dfs.update({rpm_normal : final_df})
final_dfs
```
